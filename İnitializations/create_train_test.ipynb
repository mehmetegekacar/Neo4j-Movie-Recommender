```python
import pandas as pd
import numpy as np
```

```python
ratings_selected = pd.read_csv('data\\ratings_selected.csv')
ratings_selected
```

```python
ratings_selected.sort_values(by="timestamp", inplace=True, ascending=True)
```

```python
user_ids = ratings_selected["userId"].unique()
```

```python
empty_df_train = pd.DataFrame(columns=["userId", "movieId", "rating", "timestamp"])
empty_df_test = pd.DataFrame(columns=["userId", "movieId", "rating", "timestamp"])

for i in user_ids:
    df = ratings_selected[ratings_selected["userId"] == i]
    df.sort_values(by="timestamp", inplace=True, ascending=True)
    df_train = df[:int(len(df)*0.8)]
    df_test = df[int(len(df)*0.8):]
    empty_df_train = pd.concat([empty_df_train, df_train])
    empty_df_test = pd.concat([empty_df_test, df_test])
```

```python
empty_df_train.to_csv("data\\ratings_train.csv", index=False)
```

```python
empty_df_test.to_csv("data\\ratings_test.csv", index=False)
```
